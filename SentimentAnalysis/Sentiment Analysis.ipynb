```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud, ImageColorGenerator
from nltk import sent_tokenize, pos_tag, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords, wordnet
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
data = pd.read_csv('reviews.csv' , delimiter=',')
stop = set(stopwords.words('english'))
data_bad = data[data['Score'] <= 3]
data_good = data[data['Score'] > 3]
# print(data_good)
```

```python
data = data[data['Text'].notnull()]
list_process = data['Text'].tolist()
list_good=data_good['Text'].tolist()
list_bad=data_bad['Text'].tolist()
# print(list_good)
```

```python
def remove_punctuation(input_string):
        punctuation = ".,:;!?,<>=-)(][|"
        for item in punctuation:
            for i in range(len(input_string)):
                input_string[i] = str(input_string[i]).replace(item, '')
```

```python
remove_punctuation(list_good)
remove_punctuation(list_bad)
# print(list_good)
```

```python
list_good1,list_bad1,list_process1=[],[],[]
for i in range(len(list_good)):
    stop_free = " ".join([i for i in list_good[i].lower().split() if i not in stop])
    list_good1.append(stop_free)
#     print(list_good1)
for i in range(len(list_bad)):
        stop_free = " ".join([i for i in list_bad[i].lower().split() if i not in stop])
        list_bad1.append(stop_free)
```
