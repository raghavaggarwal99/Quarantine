```python
df=pd.read_csv("electronics-INDIA.csv")
product="Vivo U20 (Blazing Blue, Snapdragon 675 AIE, 4GB RAM, 64GB Storage)"
# print(df)
category=df.loc[df['Product_Title'] == product, 'Product_Category'].iloc[0]
subcategory=df.loc[df['Product_Title'] == product, 'Product_Subcategory1'].iloc[0]
print(category)
print(subcategory)

features=['Product_Price','Product_Company','Product_Ratings','Product_Sold_By']

df = df[df['Product_Category']==category]
df = df[df['Product_Subcategory1']==subcategory]
# print(df)
```

```python
def combine_features(row):
    return row['Product_Price'].map(str) +" "+row['Product_Company']+" "+row["Product_Ratings"].map(str)+" "+row["Product_Sold_By"]
```

```python
for feature in features:
    df[feature] = df[feature].fillna('')


df["combined_features"] = df.Product_Price.map(str)+" "+df.Product_Company+" "+df.Product_Ratings.map(str)+" "+df.Product_Sold_By
length=len(df["combined_features"])
# print(df["Product_Title"])

test_features =  list(enumerate(df["combined_features"]))

similar_products =  list(enumerate(df["Product_Title"]))

# for i in range(0,length):
#     print(test_features[i][1])
indexofproduct=-1

for i in range(0,length):
#     print(similar_products[i][1])
    if(similar_products[i][1]==product):
        indexofproduct=i
        break
        

print(indexofproduct)
# print(similar_products[indexofproduct][1])
```

```python
def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return len(s1.intersection(s2)) / len(s1.union(s2))
    
main_list = test_features[indexofproduct][1].split()
# print(main_list)


rows, cols = (2, length) 
solution=[] 


for i in range(0,length):
    col = [] 
    if(i==indexofproduct):
        continue
    else:   
        score=jaccard_similarity(main_list,test_features[i][1].split())
        col.append(score) 
        col.append(similar_products[i][1]) 
        solution.append(col) 
#         print(i,score,similar_products[i][1])


solution = sorted(solution,key=lambda x:x[0],reverse=True)   

for i in range(0, len(solution)):
    s=solution[i][1]
    if(s is np.nan):
        continue
    else:
        print(s)
#     if(np.isnan(solution[i][1])):
#         continue
#     else:
#         print(solution[i][1])
    
    

```
